## Covid-19
2019 data in balkan countries

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('worldometer_data.csv')

```

```python
balkan_countries=['Greece','Bulgaria','Romania','Serbia','Croatia','Slovenia','Turkey','Albania']
balkan_x = data[data['Country/Region'].isin(balkan_countries)]
grouped_x = balkan_x.groupby('Country/Region')[['TotalCases', 'TotalDeaths', 'TotalRecovered']].mean()
data=data.dropna()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
grouped_x.plot(kind='bar', ax=ax)
ax.set_title('Total Number of Cases, Deaths, and Recoveries in the Balkan Countries')
ax.set_xlabel('Country')
ax.set_ylabel('Number of Cases')
plt.show()
```

```python
deaths_per_country = balkan_x.groupby('Country/Region')['TotalDeaths'].sum()
plt.pie(deaths_per_country, labels=deaths_per_country.index, autopct='%1.1f%%')
plt.title('Distribution of COVID-19 Deaths in the Balkans')
plt.show()
```

```python
recoveries_per_country = balkan_x.groupby('Country/Region')['TotalRecovered'].sum()
plt.pie(recoveries_per_country, labels=deaths_per_country.index, autopct='%1.1f%%')
plt.title('Distribution of COVID-19 Total Recovered in the Balkans')
plt.show()
```

```python
population_per_country = balkan_x.groupby('Country/Region')['Population'].sum()
plt.pie(population_per_country, labels=population_per_country.index, autopct='%1.1f%%')
plt.title('Distribution of COVID-19 Population in the Balkans')
plt.show()
```
